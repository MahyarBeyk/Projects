## Final dataset

```python
df_Final = df[df['کدتعرفه'].isin(List_Codes)]
df_Final.reset_index(drop=True, inplace=True)
df_Final.index = df_Final.index + 1
df_Final
```

# Exporting Tarefeh details from Tarefeh dataset into Final dataset

```python
df_Final['ماه']=''
df_Final['ماه'].to_list()
```

```python
df_Tarefeh[df_Tarefeh['کدتعرفه']==40111090]['نوع'].to_list()[0]
```

```python
df_Tarefeh[df_Tarefeh['کدتعرفه']==40111090].drop(columns=['کدتعرفه', 'نام گروه مواد و محصول', 'شرح تعرفه'])
```

```python
df_Tarefeh[df_Tarefeh['کدتعرفه']==40111090].drop(columns=['کدتعرفه', 'نام گروه مواد و محصول', 'شرح تعرفه']).columns
```

```python
for code in List_Codes:
  for column in df_Tarefeh[df_Tarefeh['کدتعرفه']==code].drop(columns=['کدتعرفه', 'نام گروه مواد و محصول', 'شرح تعرفه']).columns:
    if column=='نوع':
      df_Final.loc[df_Final['کدتعرفه']==code, 'گروه محصول']=df_Tarefeh[df_Tarefeh['کدتعرفه']==code][column].to_list()[0]
    elif column=='رینگ یا سایز':
      df_Final.loc[df_Final['کدتعرفه']==code, 'رینگ یا سایز']=df_Tarefeh[df_Tarefeh['کدتعرفه']==code][column].to_list()[0]
    elif column=='وزن هر حلقه (کیلوگرم)':
      df_Final.loc[df_Final['کدتعرفه']==code, 'وزن هر حلقه (کیلوگرم)']=df_Tarefeh[df_Tarefeh['کدتعرفه']==code][column].to_list()[0]
    elif column=='Category':
      df_Final.loc[df_Final['کدتعرفه']==code, 'Category']=df_Tarefeh[df_Tarefeh['کدتعرفه']==code][column].to_list()[0]

df_Final
```
